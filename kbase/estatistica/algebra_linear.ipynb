```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = np.array([6,-2])
b = np.array([-4,4])
vetor = a + b
vetor
```

```python
c = np.array([a,b]) # combina os arrays
val = c.ravel() # converte de N dimensões, para 1 dimensão
val
val_max = val.max() + 4
val_min = (abs(val.max()) + 4) * -1 # pega o maior valor, soma 4 (para aumentar espaço do gráfico) e multiplica por -1
```

```python
plt.axis([val_min, val_max, val_min, val_max]) # x inicial, x final, y inicial, y final
plt.grid(axis='both', linestyle='-.')
plt.plot(a, c='#0080f1', marker='s', ms=6)
plt.plot(b, c='#998022', marker='^', ms=6)
plt.show()
```
