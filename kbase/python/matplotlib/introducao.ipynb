```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-notebook')
import numpy as np
import pandas as pd
```

O Matplotlib gera um gráfico de duas dimensões, onde na parte de baixo fica o **eixo X**, a esquerda o **eixo Y** e, entre eles, nossos dados.

Para nossos exemplos a partir de agora, utilizaremos esse conjunto de dados. Nosso eixo X, por exemplo, será composto por números de 1 a 10, enquanto nosso eixo Y terá os números de 11 a 20.

De forma mais técnica, ele trabalha com o primeiro quadrante do plano cartesiano.

```python
x = np.arange(1,10)
y = np.arange(11,20)
print('Eixo x => ', x)
print('Eixo y =>', y)
```

Para verificar os temas para os gráficos, para utilizar o comando `plt.style.available`. A referência encontra-se no site do matplotlib - [link](https://matplotlib.org/devdocs/gallery/style_sheets/style_sheets_reference.html).

Para aplicar o tema, basta utilizar a função `plt.style.use('nome_tema')`.

Os gráficos gerados no Matplotlib são pequenos. Para resolver esse problema, eu gosto sempre de deixar um tamanho padrão para todos os gráficos, sem que eu tenha que configurá-los manualmente. Para isso, nós utilizarmos a função `rcParams`. Ele não só permite que você aumente o tamanho do seu gráfico, mas também mudar qualquer configuração padrão do Matplotlib. Veja como ficaria caso eu quisesse deixar o tamanho de todos os meus gráficos com 11 de largura e 7 de altura:

```python
plt.rcParams['figure.figsize'] = (11,7)
```

## Criando um gráfico

```python
plt.rcParams['figure.figsize'] = (6,4)
fig,ax = plt.subplots()
ax.plot(x, y) # usar os dados
plt.title('Título do gráfico')

plt.xlabel('EIXO X')
plt.ylabel('EIXO Y')

plt.show() # mostrar o gráfico
```

### Matplotlib + Pandas
Criar um DataFrame

```python
df = pd.DataFrame({
    'X': np.arange(1,10),
    'Y': np.arange(11,20)
})
df
```

```python
df.plot()
plt.title('DF com Pandas')

plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')

plt.show()
```

```python
chart = df[['X','Y']].plot(kind='bar', title='Título', figsize=(10,6), legend=True, fontsize=10)
chart.set_xlabel("Hour", fontsize=12)
chart.set_ylabel("V", fontsize=12)
plt.show()
```

```python
df.plot() #criando gráfico
plt.title('SEU TÍTULO LINDO') #adicionando o título
plt.xlabel('NOME DO EIXO X') #definindo nome do eixo X
plt.ylabel('NOME DO EIXO Y') #definindo nome do eixo Y
plt.legend(loc='best') #colocando a legenda no melhor lugar
plt.show() #mostrando gráfico
```

**Dois gráficos em um**

```python
plt.plot(x, y, 'b--')
plt.scatter(x, y, marker="*", color='red')
plt.show()
```

**Gráficos com anotações**

```python
fig, ax = plt.subplots()
ax.bar(x, y)
ax.annotate("Maior valor", 
            xy=(9, 16),
            xycoords='data',
            xytext=(11, 10),
            textcoords='data',
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3"))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

ax.set_ylim(-2,2)
plt.show()
```

# Scatter Star Poly


Create multiple scatter plots with different star symbols. [link](https://matplotlib.org/gallery/lines_bars_and_markers/scatter_star_poly.html#sphx-glr-gallery-lines-bars-and-markers-scatter-star-poly-py)

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


x = np.random.rand(10)
y = np.random.rand(10)
z = np.sqrt(x**2 + y**2)

plt.subplot(321)
plt.scatter(x, y, s=80, c=z, marker=">")

plt.subplot(322)
plt.scatter(x, y, s=80, c=z, marker=(5, 0))

verts = np.array([[-1, -1], [1, -1], [1, 1], [-1, -1]])
plt.subplot(323)
plt.scatter(x, y, s=80, c=z, marker=verts)

plt.subplot(324)
plt.scatter(x, y, s=80, c=z, marker=(5, 1))

plt.subplot(325)
plt.scatter(x, y, s=80, c=z, marker='+')

plt.subplot(326)
plt.scatter(x, y, s=80, c=z, marker=(5, 2))

plt.show()
```
