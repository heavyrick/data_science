### histograma
Um histograma é uma ferramenta de análise e representação de dados quantitativos, agrupados em classes de freqüência que permite distinguir a forma, o ponto central e a variação da distribuição, além de outros dados como amplitude e simetria na distribuição dos dados. [Brasil Escola](https://www.infoescola.com/estatistica/histograma/)

```python
fig, ax = plt.subplots()
plt.hist(estados['total'], bins=15)
#estados['total'].hist()
plt.show()
```

```python
fig, ax = plt.subplots()
plt.ylim(estados['total'].min(),estados['total'].max())
plt.hist(estados['total'], bins=10, orientation='horizontal')
ax.ticklabel_format(style='plain')
plt.show()
```

```python
# criar coluna percentual
estados['percent'] = estados['total'] / estados['total'].sum()
estados
```

```python
plt.pie(estados['percent'], labels=estados.index, autopct='%1.2f%%')
plt.show()
```

## Iris

```python
pydataset.data('iris', show_doc=True)
```

```python
iris = pydataset.data('iris')
iris.head()
#iris.tail()
```

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'], sizes=50 * iris['Petal.Length'])
```

```python
def specie_color(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    return 2
```

```python
iris['SpeciesNumber'] = iris['Species'].apply(specie_color)
```

```python
plt.scatter(
    iris['Sepal.Length'], iris['Sepal.Width'], sizes=20 * iris['Petal.Length'],
    c=iris['SpeciesNumber'], cmap='viridis', alpha=0.4
)
```

## Base de clientes
Clientes das cidades do ddd 19

```python
clientes = pd.read_csv('data/clientes_ddd19_geodados.csv', delimiter=';')
clientes.head()
```

```python
# Selecionando as latitudes e longitudes da cidade de Americana
cli_americana = clientes[clientes['city_name'] == 'AMERICANA'][['latitude','longitude','id','route_name', 'status']]
cli_americana
```

```python
plt.scatter(cli_americana['longitude'], cli_americana['latitude'], marker='.')
```

O problema de falta de uniformidade dos dados, é porque os dados registrados estão errado. Muitas geolocalizações estão incorretas. Ao plotar os dados no mapa é possível ver isso. Para corrigir, podemos eliminar alguns dados que não nos interessam.

> Limites de Americana:
* Oeste: -22.697980, -47.387655
* Sul: -22.784127, -47.350949
* Leste:-22.693533, -47.216712
* Norte: -22.679940, -47.363276

Vamos considerar apenas o maior e o menor valor de latitude e longitude. Lembrando que a referência da latitude é a Linha do Equador (e a leitura se faz entre Norte-Sul), e a referência da longitude é o Meridiano de Greenwich (e a leitura se faz Leste-Oeste).

**Latitude**
* max (N): -22.679940
* min (S): -22.784127

**Longitude**
* max (L): -47.216712
* min (O): -47.387655

A ideia é pegarmos os clientes que se encaixam dentro desses limites, e deixarmos os outros que estão incorretos de fora.

```python
lat_norte = -22.679940
lat_sul = -22.784127
lng_leste = -47.216712
lng_oeste = -47.387655
```

```python
cli_americana_ativo = cli_americana[
    (cli_americana['status'] == 1) &
    ((cli_americana['latitude']>=lat_sul) & (cli_americana['latitude']<=lat_norte)) &
    ((cli_americana['longitude']>=lng_oeste) & (cli_americana['longitude']<=lng_leste))
]
cli_americana_ativo.describe()
```

```python
cli_americana_inativo = cli_americana[
    (cli_americana['status'] == 0) &
    ((cli_americana['latitude']>=lat_sul) & (cli_americana['latitude']<=lat_norte)) &
    ((cli_americana['longitude']>=lng_oeste) & (cli_americana['longitude']<=lng_leste))
]
cli_americana_inativo.describe()
```

```python
# Agora que retiramos as coordenadas muito erradas, podemos repetir a operação
plt.scatter(cli_americana_inativo['longitude'], cli_americana_inativo['latitude'], marker='.',  c='blue')
plt.scatter(cli_americana_ativo['longitude'], cli_americana_ativo['latitude'], marker='.',  c='red')
```

```python
import mplleaflet
plt.scatter(cli_americana_inativo['longitude'], cli_americana_inativo['latitude'], marker='.',  c='blue')
plt.scatter(cli_americana_ativo['longitude'], cli_americana_ativo['latitude'], marker='.',  c='red')
mplleaflet.show()
```
