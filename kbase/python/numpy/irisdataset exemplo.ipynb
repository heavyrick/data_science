### iris dataset
[Machine learning repository](https://archive.ics.uci.edu/ml/datasets/Iris)

1. sepal length in cm
2. sepal width in cm
3. petal length in cm
4. petal width in cm
5. class: 
  * Iris Setosa
  * Iris Versicolour
  * Iris Virginica
  
São 150 registros, com 50 valores para cada classe de pétalas.

```python
import numpy as np
```

```python
data = np.genfromtxt("data/iris.data.txt", delimiter=",", usecols=(0,1,2,3))
data
```

```python
len(data) # nº linhas
```

## 1 - Trabalhando com a coluna ´sepal length´

```python
data[:50,0] # dados do comprimento da iris setosa (1º coluna), que são as 50 primeiras linhas
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Plotando as `sepal length` de **Iris-setosa**

```python
plt.plot(data[:50,0], c='Red', ls=':', marker='s', ms=8)
plt.show()
```

Plotando as `sepal length` de **Iris-versicolor**

```python
plt.plot(data[51:100,0], c='Green', ls=':', marker='s', ms=8)
plt.show()
```

Plotando as `sepal length` de **Iris-virginica**

```python
plt.plot(data[101:,0], c='Blue', ls=':', marker='s', ms=8)
plt.show()
```

Plotando as `sepal length` de todas as sépalas

```python
plt.plot(data[:50,1], c='Red', ls=':', marker='s', ms=8, label='Iris-setosa')
plt.plot(data[51:100,1], c='Green', ls=':', marker='o', ms=8, label='Iris-versicolor')
plt.plot(data[101:,1], c='Blue', ls=':', marker='*', ms=8, label='Iris-virginica')
plt.legend()
plt.show()
```

Observando o gráfico acima, podemos verificar que a **iris-setosa** tem as pétalas com maior comprimento em centímetros, seguida pela **iris-virginica**, e com pétalas menores a **iris-versicolor**.
