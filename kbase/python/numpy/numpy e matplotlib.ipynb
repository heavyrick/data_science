```python
# https://matplotlib.org/api/pyplot_summary.html

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([1200,1500,1300,1400,1100])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c='Green', ls='--', marker='s', ms=5) # c = cor, ls=estilo da linha, marker=marcador, ms=tamanho do marcador
plt.show()
```

```python
# Pontos dos jogadores da NBA, da temporada 2013-2014 até 2016-2017
# link: http://stats.nba.com/player/201566/career/?PerMode=Totals

lebron_james = np.array([2089, 1743, 1920, 1954])
stephen_curry = np.array([1873, 1900, 2375, 2639])
russel_westbrook = np.array([1002, 1886, 1878, 2558])
isaiah_thomas = np.array([1465, 2202, 1823, 2199])
anthony_davis = np.array([1394, 1656, 1481, 2099])
```

```python
# https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

plt.plot(lebron_james, c='Blue', marker='s', ms=6, label='Lebron James')
plt.plot(stephen_curry, c='Red', marker='^', ms=6, label='Stephen Curry')
plt.plot(russel_westbrook, c='Brown', marker='o', ms=6, label='Russel Westbrook')
plt.plot(isaiah_thomas, c='Green', marker='*', ms=6, label='Isaiah Thomas')
plt.plot(anthony_davis, c='Orange', marker='>', ms=6, label='Anthony Davis')
plt.legend(loc='lower right')
plt.show()
```
